```python
import pandas as pd
table1 = pd.read_csv('3-Table1.csv')
table2 = pd.read_csv('3-Table2.csv')
```

```python
df_master = pd.merge(table1, table2, left_on='userid', right_on='userid', how='left')
```

```python
df_master
```

```python
df_master['fail'] = df_master.apply(lambda x: 1 if x['success']==0 else 0, axis=1)
```

```python
df_master
```

```python
gk = df_master.groupby('data_center')
```

```python
gk['fail'].sum() / gk['fail'].count()
```

```python
gk = df_master.groupby('country')
```

```python
gk['fail'].sum() / gk['fail'].count()
```

```python
failed = df_master[df_master['fail'] == 1]['userid'].unique()
```

```python
df_master_never_failed = df_master[~df_master['userid'].isin(failed)]
```

```python
df_master_never_failed
```

```python
gk = df_master_never_failed.groupby('country')
```
